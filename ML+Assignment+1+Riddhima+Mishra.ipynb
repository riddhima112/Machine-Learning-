C. Now assume that you donâ€™t have access to all the raw data and you know the following things: 
 
There are a total of 1000 calls, 600 from Garfield Park and 400 from Uptown. Of the 600 calls from Garfield Park, 100 of them are about Graffiti Removal. Of the 400 calls from Uptown, 160 are about Graffiti Removal. If a call comes about Graffiti Removal, how much more/less likely is it that the call came from Garfield Park versus Uptown? 


For answering this question, we can assume the events to be independent and can calculate their conditional probability. 

(Prob(call from garfield | graffiti call) = P(call from garfield) * P(graffiti call from garfield) = (600/1000)*(100/600) = 10 percent

Prob(call from Uptown | graffiti call) = P(call from uptown) * P (graffiti call from Uptown) = (400/1000)*(160/400) = 16 percent
